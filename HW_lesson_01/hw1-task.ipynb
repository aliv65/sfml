```python
# (4) Сгрупировать по "default payment next month" и посчитать медиану для всех показателей начинающихся на BILL_ и PAY_
filterCondition = df.columns.str.startswith('BILL_') | df.columns.str.startswith('PAY_') | df.columns.str.startswith('default.')
df.loc[:, filterCondition].groupby('default.payment.next.month').median()
```

```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

def limitBalDescription(limitBal):
    if limitBal <= 10000:
        return 'A'
    elif limitBal > 10000 and limitBal <= 100000:
        return 'B'
    elif limitBal > 100000 and limitBal <= 200000:
        return 'C'
    elif limitBal > 200000 and limitBal <= 400000:
        return 'D'
    elif limitBal > 400000 and limitBal <= 700000:
        return 'E'
    elif limitBal > 700000:
        return 'F'
    else:
        return 'Not determined'

df['Col'] = df['LIMIT_BAL'].map(lambda i: limitBalDescription(i))
df.head()
```
